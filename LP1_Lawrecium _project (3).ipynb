#### Startup Stages

Startups typically progress through several stages of funding rounds, often referred to as "series." These series are indicative of the company's growth and development. Here are the common types of startup stages or series:


1. **Seed Stage (Pre-Seed, Seed Round):**
   - This is the earliest stage of funding where the startup is just getting off the ground.
   - Seed funding is used to validate the business idea, develop a prototype, and conduct initial market research.
   - Investors at this stage could include friends, family, angel investors, or early-stage venture capital firms.

2. **Series A:**
   - At this stage, the startup has usually developed a product or service and has some initial traction in the market.
   - Series A funding is typically used to scale the business operations, expand the team, and further develop the product.
   - Venture capital firms are the primary investors in Series A rounds, although some angel investors and early-stage investors may also participate.

3. **Series B:**
   - By the Series B stage, the startup has demonstrated significant growth potential and has a solid customer base.
   - The funding raised in Series B rounds is often used to further scale the business, enter new markets, and invest in marketing and sales efforts.
   - Venture capital firms continue to be the main source of funding at this stage, along with some participation from strategic investors.

4. **Series C and Beyond:**
   - These stages represent subsequent rounds of funding as the startup continues to grow and expand.
   - Series C, D, E, and so forth rounds may involve larger amounts of funding and can be used for various purposes such as further expansion, acquisitions, or preparing for an initial public offering (IPO).
   - At these later stages, the startup may attract investments from larger venture capital firms, private equity firms, hedge funds, or even corporate investors.

5. **Bridge Rounds:**
   - In between major funding rounds, startups may sometimes raise bridge rounds to provide additional runway while they prepare for the next significant financing round.
   - Bridge rounds are often smaller and may involve existing investors providing additional capital to sustain operations until the next major round.

# Hypothesis Tests


# 1

#### H0 - There is no significant difference in the average funding received for bangalore and Mumbai in the year 2020
#### H1 - There is significant difference in the average funding received for Bangalore and Mumbai in the year 2020


```python
# Filter for Bangalore and Mumbai funding amounts for 2020
funding_Bengaluru_2020 = df_concat[(df_concat['Location'] == 'Bengaluru') & (df_concat['Year'] == 2020)]['Amount(1000)']
funding_mumbai_2020 = df_concat[(df_concat['Location'] == 'Mumbai') & (df_concat['Year'] == 2020)]['Amount(1000)']
```

```python
# Function to calculate mean difference between samples
def mean_diff(sample1, sample2):
    return np.mean(sample1) - np.mean(sample2)
```

```python
num_iterations = 1000  # Number of bootstrap iterations
mean_diffs = []

for _ in range(num_iterations):
    # Randomly sample with replacement from Bangalore and Mumbai for 2020
    sample_Bengaluru = df_concat[(df_concat['Year'] == 2020) & (df_concat['Location'] == 'Bengaluru')]['Amount(1000)']
    sample_mumbai = df_concat[(df_concat['Year'] == 2020) & (df_concat['Location'] == 'Mumbai')]['Amount(1000)']
    
    # Calculate mean difference
    diff = mean_diff(sample_Bengaluru, sample_mumbai)
    mean_diffs.append(diff)


# Calculate p-value based on the mean differences
observed_diff = mean_diff(funding_Bengaluru_2020, funding_mumbai_2020)

p_value_bootstrap = np.sum(np.abs(mean_diffs) >= np.abs(observed_diff)) / num_iterations

# Print the results
print("Bootstrap Hypothesis Testing:")
print("Observed Mean Difference:", observed_diff)
print("P-value (Bootstrap):", p_value_bootstrap)

# Check if p-value is less than significance level (e.g., 0.05)
if p_value_bootstrap < 0.05:
    print("Reject null hypothesis: Average funding in 2020 is significantly different between Bangalore and Mumbai.")
else:
    print("Fail to reject null hypothesis: Average funding in 2020 is not significantly different between Bangalore and Mumbai.")
```

## Observation 
The observed mean difference of 1343.27 indicates that, on average, the funding amount in Bangalore for the year 2020 is higher than the funding amount in Mumbai by approximately 1343.27 units (the units depend on the currency or scale of your data).

However, the p-value obtained from the Bootstrap Hypothesis Testing is 1.0. A p-value of 1.0 indicates that we fail to reject the null hypothesis. In this context, the null hypothesis might be that there is no difference in the average funding amount between Bangalore and Mumbai for the year 2020.

Since we fail to reject the null hypothesis, we do not have enough evidence to conclude that the average funding amount in 2020 is significantly different between Bangalore and Mumbai. This means that despite the observed mean difference, the difference is not statistically significant based on the data and analysis performed.

# 2


#### H0 - There is no significant difference between the sample mean and population mean for the average amount of funding received in Bangalore
#### H1 - There is significant difference between the sample mean and population mean for the average amount of funding received in Bangalore


```python
import numpy as np
import pandas as pd
from scipy import stats



# Bootstrap sampling
bootstrap = []
num_iterations = 500

for i in range(num_iterations):
    # Sample with replacement from Mumbai funding data and calculate mean
    sample_mean = np.mean(df_concat[df_concat['Location'] == 'Bengaluru']['Amount(1000)'].sample(frac=1, replace=True))
    bootstrap.append(sample_mean)
```

```python
# Calculate observed mean of Mumbai funding amounts
observed_mean = df_concat[df_concat['Location'] == 'Bengaluru']['Amount(1000)'].mean()

# Calculate the difference between observed mean and bootstrap mean
bootstrap_mean = np.mean(bootstrap)
mean_diff = observed_mean - bootstrap_mean

# Calculate standard error of the mean for the bootstrap distribution
std_error = np.std(bootstrap, ddof=1) / np.sqrt(len(bootstrap))

# Calculate t-statistic
t_stat = mean_diff / std_error

# Calculate p-value
degrees_of_freedom = len(bootstrap) - 1
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=degrees_of_freedom))

# Print the results
print("One-Sample T-Test Results:")
print("Observed Mean of Mumbai Funding:", observed_mean)
print("Mean of Bootstrap Distribution:", bootstrap_mean)
print("Difference in Means:", mean_diff)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

## Observation


However, the p-value obtained from the Bootstrap Hypothesis Testing is 0.99 which is greater than alpha. A p-value of 0.99 indicates that we fail to reject the null hypothesis. In this context, the null hypothesis might be that there is no difference in the sample mean and population mean.
